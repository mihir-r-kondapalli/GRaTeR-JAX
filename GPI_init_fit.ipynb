```python
from astropy.stats import bayesian_info_criterion as bic
from astropy.stats import akaike_info_criterion as aic

def delta_criteria(ll1,np1,ll2,np2):
    d_bic = bic(ll1,np1,32400)-bic(ll2,np2,32400)
    d_aic = aic(ll1,np1,32400)-aic(ll2,np2,32400)
    print('$\Delta$BIC is',d_bic)
    print('$\Delta$AIC is',d_bic)
    print('If delta criteria are negative, the first model is preferred')
```
