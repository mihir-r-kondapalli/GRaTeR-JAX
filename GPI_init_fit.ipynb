```python
from astropy.stats import bayesian_info_criterion as bic
from astropy.stats import akaike_info_criterion as aic

def delta_criteria(ll1,np1,ll2,np2):
    d_bic = bic(ll1,np1,32400)-bic(ll2,np2,32400)
    d_aic = aic(ll1,np1,32400)-aic(ll2,np2,32400)
    print('$\Delta$BIC is',d_bic)
    print('$\Delta$AIC is',d_bic)
    print('If delta criteria are negative, the first model is preferred')
```

## HR 4796A

```python
model1, params1, ll1, spf1 = gpi.init_fit('hr4796a_H_pol',num_knots=9,save=True)
```

```python
import statistical_analysis.interpolated_univariate_spline as spline_func
import numpy as np

hi = np.rad2deg(np.arccos(spf1['low_bound']))
lo = np.rad2deg(np.arccos(spf1['up_bound']))
print(hi,lo)

spf = spline_func.InterpolatedUnivariateSpline(np.linspace(lo,hi,9),spf1['knot_values'])

func = spf.__call__(np.linspace(0,180,100))

import matplotlib.pyplot as plt
plt.plot(np.linspace(0,180,100),func)
plt.scatter(np.linspace(lo,hi,9),spf1['knot_values'],color='orange')

plt.title('HR 4796A Pol SPF Initial Fit (9 knots)')
plt.xlabel('Scattering Angle (degrees)')
plt.ylabel('Arbitrary Units')
plt.savefig('../GPI_results/best_fits/hr4796a_pol_spf_init.png',bbox_inches='tight')
```

```python
##best model for HR 4796 has **9 knots**
```
